```python
fig, ax = plt.subplots(2,2,figsize=(11,11))

ax[0][0].set_frame_on(False)
model.plot_continuous_links(ax=ax[0][0], parameter = 'velocity', value='max', unit='ft/s', 
                            link_width=2, link_style='-', include_pumps=True, include_valves=True, draw_tanks=True, draw_reservoirs=True, draw_pumps=True, draw_valves=True, draw_links=True, 
                            base_legend_loc='lower left', base_legend_label_font_size=8, 
                            reservoir_size=50, tank_size=50, valve_size=75, pump_width=2, pump_line_style='--')

ax[0][1].set_frame_on(False)
model.plot_continuous_links(ax=ax[0][1], parameter = 'flowrate', value=10, cmap='cool', color_bar_title='Flowrate [m^3/s] at Timestep Ten', 
                            base_legend_loc='lower left', base_legend_label_font_size=8,base_legend_label_color='b', draw_legend_frame=True, 
                            reservoir_size=50, reservoir_color='r', reservoir_shape='v', reservoir_border_color='k', reservoir_border_width=1, 
                            tank_size=50, tank_color='g', tank_shape='d', tank_border_color='k', tank_border_width=1, 
                            valve_element='node', valve_size=75, valve_color='m', valve_shape='*', valve_border_color='k', valve_border_width=1)

ax[1][0].set_frame_on(False)
model.plot_continuous_links(ax=ax[1][0], parameter='headloss', value='mean', cmap='spring', 
                            base_legend_loc='lower left', base_legend_label_font_size=8, 
                            reservoir_size=50, tank_size=50, valve_size=75, 
                            pump_element='node', pump_color='g', pump_shape='*', pump_size=50, pump_border_color='k', pump_border_width=1, 
                            draw_nodes=False)

ax[1][1].set_frame_on(False)
model.plot_continuous_links(ax=ax[1][1], parameter='quality', value='mean', unit='hr',
                            base_legend_loc='lower left', base_legend_label_font_size=8,
                            reservoir_size=50, tank_size=50,
                            vmin=20, vmax=70, color_bar_width=0.02, color_bar_height=0.9, 
                            valve_element='link', valve_width=2, valve_line_style='-')
```
